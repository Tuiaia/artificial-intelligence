# News of the Brazilian Newspaper dataset
link: https://www.kaggle.com/datasets/marlesson/news-of-the-site-folhauol?resource=download

```python
raw_articles_df = pd.read_csv('articles.csv')
```

```python
# Visualizando as dimensões do df
raw_articles_df.shape
```

```python
# Visualizando os cinco primeiros exemplos do df
raw_articles_df[:5]
```

```python
# Removendo colunas que não serão utilizadas
articles_df = raw_articles_df.drop(columns=['date', 'subcategory', 'link'])
articles_df[:5]
```

```python
# Filtrando notícias relacionadas ao mercado
market_articles_df = articles_df[articles_df['category'] == 'mercado']
market_articles_df[:5]
```

```python
# Visualizando a quantidade de notícias relacionadas ao mercado
market_articles_df.shape
```

```python
# Resetando os índices do df
market_articles_df.reset_index(inplace=True, drop=True)
market_articles_df[:5]
```

```python
# Removendo coluna de 'categoria'
market_articles_df.drop(columns=['category'], inplace=True)
market_articles_df[:5]
```

```python
# Concatenando o título com o corpo da notícia e removendo as colunas individuais
market_articles_df.loc[:, 'concat_title_text'] = market_articles_df.loc[:, 'title'] + '. ' + market_articles_df.loc[:, 'text']
market_articles_df.drop(columns=['title', 'text'], inplace=True)
market_articles_df.columns = ['text']
```

# Stock Market News Data in Portuguese
link: https://www.kaggle.com/datasets/mateuspicanco/financial-phrase-bank-portuguese-translation

```python
raw_smn_df = pd.read_csv('financial_phrase_bank_pt_br.csv')
```

```python
# Visualizando as dimensões do df
raw_smn_df.shape
```

```python
# Visualizando a quantidade de exemplos de cada classe
raw_smn_df.loc[:, 'y'].value_counts()
```

```python
raw_smn_df[:5]
```

```python
# Copiando df para evitar novos carregamentos após modificações
smn_df = raw_smn_df.copy()
```

```python
# Invertendo a ordem das colunas para visualização 'exemplo: classe'
smn_df = smn_df[['text_pt', 'y']]
smn_df.columns = ['text', 'y']
smn_df
```

```python
def encode_classes(class_name):
    if class_name == 'positive':
        return 2
    if class_name == 'neutral':
        return 1
    if class_name == 'negative':
        return 0
```

```python
smn_df['y'] = smn_df['y'].apply(encode_classes)
smn_df
```

```python
smn_df = smn_df.drop_duplicates(subset='text')
smn_df
```

```python
smn_df.to_csv('./stock_market_news_labelled.csv', index=False)
```
