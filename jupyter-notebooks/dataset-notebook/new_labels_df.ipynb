```python
import json

def save_as_json(df, output_file):
    converted_data = []

    for index, row in df.iterrows():
        item = {
            'id': index,
            'data': {
                'text': row['text'],
                'raw_text': row['raw_text'],
                'lang': row['lang'],
                'font': row['font'],
                'sentiment': row['sentiment'],
                'impact_intensity': row['impact_intensity'],
                'impact_term': row['impact_term']
            }
        }
        converted_data.append(item)

    with open(output_file, 'w', encoding='utf-8') as outfile:
        json.dump(converted_data, outfile, ensure_ascii=False, indent=4)
```

```python
def simplify_label(label):
    label = label.lower()
    keywords = {
        'alto': 'alto',
        'baixo': 'baixo',
        'neutro': 'neutro',
        'curto': 'curto',
        'longo': 'longo',
        'high': 'alto',
        'low': 'baixo',
        'neutral': 'neutro',
        'short': 'curto',
        'long': 'longo'
    }
    label_split = set(label.split())
    for keyword in keywords.keys():
        if keyword in label_split:
            return keywords[keyword]
```

```python
df = pd.read_excel('datasets/df-v2/df-eng-fold-1.xlsx')
df
```

```python
df = df[df['lang']=='eng'].dropna(subset='impact_intensity').reset_index(drop=True)
df
```

```python
df['impact_intensity'] = df['impact_intensity'].apply(simplify_label)
df
```

```python
df['impact_term'] = df.apply(lambda x: 'neutral' if x['impact_intensity'] == 'neutral' else None, axis=1)
df
```

```python
df['impact_term'] = df.apply(lambda x: 'neutro' if x['impact_intensity'] == 'neutro' else None, axis=1)
df
```

```python
df = df.sort_values(by='impact_intensity').reset_index(drop=True)
df
```

```python
df.to_csv('datasets/df-v2/df-eng-fold-1.csv', sep='|', index=False)
```
