```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Português

```python
def get_news(url):
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Falha na solicitação, código de status: {response.status_code}")
    else:
        print("Solicitação bem-sucedida")

        soup = BeautifulSoup(response.content, 'html')
        h4_elements = soup.find_all('h4', {'class': 'gPFEn'})
        texts = [h4.text for h4 in h4_elements]
        
        df = pd.DataFrame(columns=['text', 'label'])
        df['text'] = texts

        return df
```

```python
url = 'https://news.google.com/topics/CAAqKggKIiRDQkFTRlFvSUwyMHZNRGx6TVdZU0JYQjBMVUpTR2dKQ1VpZ0FQAQ?hl=pt-BR&gl=BR&ceid=BR%3Apt-419'
```

```python
df_ptbr = get_news(url)
```

```python
df_ptbr
```

```python
df_ptbr.to_csv('google-news-ptbr.csv', sep='|')
```
